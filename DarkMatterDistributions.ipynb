```python
df['LumCalc'] = df['Reff'] ** 2 * 2 * np.pi * df['SBeff'] / (10 ** 3)
df['LumCalc'] = df["L[3.6]"] / ((1 + np.e ** ((-1 * df['Reff'] * df['SBeff']))))
df["LumCalc"]
```

Formula: $L=\frac{\pi R_{eff}^2 S}{500}$<br>
Where $L$ is luminosity in $L_\odot$, $R_{eff}$ is the effective radius in kpc, and $S$ is the effective surface brightness in $L_\odot/pc^2$

```python
df['LumPercentError'] = np.abs(df['L[3.6]'] - df['LumCalc']) / df['L[3.6]']
```

```python
q1_avg = np.mean(df[df['Q'] == 1]['LumPercentError'])
q2_avg = np.mean(df[df['Q'] == 2]['LumPercentError'])
q3_avg = np.mean(df[df['Q'] == 3]['LumPercentError'])
print(q1_avg, q2_avg, q3_avg)
print(np.mean(df['LumPercentError']))
```

```python
# Get only the rows with these galaxies: NGC3741, NGC4138, UGC06614, NGC2915, NGC4278
galaxies = ["NGC2915", "NGC2998", "NGC3741", "NGC4138", "UGC00191", "UGC06614", "NGC4278"]
galaxy_legend = ["NGC2915 (Irregular)", "NGC2998 (Middle Spiral)", "NGC3741 (Irregular)", "NGC4138 (Lenticular)", "UGC00191 (Irregular)", "UGC06614 (Tight Spiral)", "NGC4278 (Elliptical)"]
df3 = df3[df3["Galaxy"].isin(galaxies)]
df3 = df3.reset_index()
df3
```

```python
data = []
new_data = {galaxy: [[], []] for galaxy in galaxies}
print(new_data)
for galaxy in galaxies:
    print(galaxy)
    fin = open("Data/Galaxy Data/Rotmod_Undivided/" + galaxy + "_rotmod.dat")
    lines = fin.read().split("\n")[3:-1]
    for line in lines:
        data.append(line.strip().replace(",", ";").split())
    for line in data:
        new_data[galaxy][0].append(float(line[0]))
        new_data[galaxy][1].append(float(line[1]))
    data.clear()
print(data)
new_data
```

```python
# append the data as two new columns to the dataframe in a column called radius and vrot for each galaxy (for the row with NGC3741 there should be a list under radius and velocity)
r = []
v = []
for galaxy in galaxies:
    # print(new_data)
    print(new_data[galaxy][0])
    r.append(new_data[galaxy][0])
    v.append(new_data[galaxy][1])
rad = pd.Series(r, name="Radius")
vel = pd.Series(v, name="Velocity")
df3 = pd.concat([df3, rad, vel], axis=1)
df3
```

```python
# Convert radii from arcseconds to parsecs

driver = webdriver.Chrome()
driver.get("http://arcsec2parsec.joseonorbe.com/index.html")

driver.find_element(By.NAME, "ASunits").send_keys("arcsec")
driver.find_element(By.NAME, "Dunits").send_keys("pc")

def convertarcsectopc(r, dis):
    angsize = driver.find_element(By.NAME, "txtAngsize")
    angsize.clear()
    angsize.send_keys(r)

    distance = driver.find_element(By.NAME, "txtDistance")
    distance.clear()
    distance.send_keys(dis)

    # br["ASunits"] = ["arcsec"]
    # br["Dunits"] = ["pc"]
    # br["LSunits"] = ["kpc"]

    driver.find_element(By.ID, "btn-convert").click()

    return driver.find_element(By.NAME, "txtLinsize").get_attribute("value")

# append the data as two new columns to the dataframe in a column called radius and vrot for each galaxy (for the row with NGC3741 there should be a list under radius and velocity)
r = []
r2 = []
mu = []
for galaxy in galaxies:
    new_df = sfb[sfb["galaxy"] == galaxy]
    rad2 = list(new_df["radius"])
    for i in range(len(rad2)):
        rad2[i] = convertarcsectopc(rad2[i], list(df3[df3["Galaxy"] == galaxy]["Dist"])[0])
    r.append(list(map(float, list(new_df["radius"]))))
    r2.append(list(map(float, rad2)))
    mu.append(list(map(float, list(new_df["mu"]))))
    print(list(new_df["radius"]))
driver.quit()
origrad = pd.Series(r, name="Original Radius (1)")
rad = pd.Series(r2, name="Radius2")
sb2 = pd.Series(mu, name="SB")
df3 = pd.concat([df3, origrad, rad, sb2], axis=1)
df3 = df3.head(7)
df3
```

```python
for i in range(len(galaxies)):
    radius = df3["Radius"].tolist()[i]
    mass = df3["Mass"].tolist()[i]
    def lum_function(r):
        global i
        current_r = df3["Radius2"][i]
        current_lum = df3["Lum2"][i]
        # find two closest points to r in current_r
        def find_closest_numbers(numbers, target):
            numbers2 = sorted(numbers, key=lambda x:abs(x-target))
            return [numbers.index(numbers2[0]), numbers.index(numbers2[1])]

        closest_index = find_closest_numbers(current_r, r)
        # Interpolate a data value for the two points
        x1 = current_r[closest_index[0]]
        x2 = current_r[closest_index[1]]
        y1 = current_lum[closest_index[0]]
        y2 = current_lum[closest_index[1]]
        m = (y2 - y1) / (x2 - x1)
        b = y1 - m * x1
        return m * r + b
    ml = [mass[j] / (lum_function(radius[j])) for j in range(len(radius))]
    plt.plot(radius, ml, label=galaxy_legend[i])
plt.legend()
plt.xlabel("Radius (kpc)")
plt.rcParams['text.usetex'] = True
plt.ylabel(r"M/L ($M_{\odot}/L_{\odot}$)")
plt.rcParams['text.usetex'] = False
plt.yscale("log")
plt.xscale("log")
plt.show()
```
